```python
import pandas as pd
import sqlite3
import os

# Load the clean master table
master = pd.read_csv("../data/cleaned/master_orders.csv")

print(f"Master table loaded ✓")
print(f"Rows:     {len(master):,}")
print(f"Columns:  {master.shape[1]}")
```

```python
# Connect to the database file
db_path = "C:/Users/USER-PC/OneDrive/Desktop/olist-ecommerce-project/olist_database.db"
conn = sqlite3.connect(db_path)

# Write master table to the database
master.to_sql(
    name='master_orders',      # this becomes the table name in the database
    con=conn,
    if_exists='replace',       # replace if it already exists
    index=False                # don't write the pandas row numbers
)

print("Data written to database ✓")
print()

# Confirm by reading it back
check = pd.read_sql("SELECT COUNT(*) as total_rows FROM master_orders", conn)
print(f"Rows in database: {check['total_rows'][0]:,}")

conn.close()
print("Connection closed ✓")
```
