```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

print("All libraries loaded ✓") 
```

```python
path = "../data/raw/"

customers   = pd.read_csv(path + "olist_customers_dataset.csv")
geo         = pd.read_csv(path + "olist_geolocation_dataset.csv")
order_items = pd.read_csv(path + "olist_order_items_dataset.csv")
payments    = pd.read_csv(path + "olist_order_payments_dataset.csv")
reviews     = pd.read_csv(path + "olist_order_reviews_dataset.csv")
orders      = pd.read_csv(path + "olist_orders_dataset.csv")
products    = pd.read_csv(path + "olist_products_dataset.csv")
sellers     = pd.read_csv(path + "olist_sellers_dataset.csv")
categories  = pd.read_csv(path + "product_category_name_translation.csv")

print("All 9 tables loaded ✓")
```

```python
tables = {
    "customers":   customers,
    "geo":         geo,
    "order_items": order_items,
    "payments":    payments,
    "reviews":     reviews,
    "orders":      orders,
    "products":    products,
    "sellers":     sellers,
    "categories":  categories,
}

for name, df in tables.items():
    print(f"{name:15s} → {df.shape[0]:,} rows   {df.shape[1]} columns")
```

```python
for name, df in tables.items():
    nulls = df.isnull().sum().sum()
    print(f"{name:15s} → {nulls:,} missing values")
```
